```python
top_10 = a.head(10)
# Splitting the 'Language' column by slashes and converting to a list of languages
language_lists = top_10['Language'].str.split('/')
# Creating a list of all languages
all_languages = []
for language_list in language_lists:
    all_languages.extend(language_list)
# Count occurrences of each language
language_counts = pd.Series(all_languages).value_counts()
# Plotting the pie chart
plt.figure(figsize=(10, 8))
# Plot pie chart
plt.pie(language_counts, labels=language_counts.index, autopct='%1.1f%%', startangle=140)
# Add title above the pie chart
plt.title('Distribution of Languages in Top 10 Movies', pad=20, fontsize=16)
# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
# Step 1: Create a new column containing the count of genres used in each movie
a['Genre_Count'] = a['Genre'].apply(lambda x: len(x.split('/')))

# Step 2: Filter the DataFrame to include only movies with 1, 2, or 3 genres
filtered_movies_genre = a[a['Genre_Count'].isin([1, 2, 3])]

# Step 3: List the movies based on the count of genres used
for count in [1, 2, 3]:
    movies_with_count_genre = filtered_movies_genre[filtered_movies_genre['Genre_Count'] == count][['Title', 'Genre']]
    if not movies_with_count_genre.empty:
        print(f"Movies with {count} genre(s):")
        # Display the table with title, genres
        print(movies_with_count_genre.to_string(index=False))
        print()
    else:
        print(f"No movies with {count} genre(s) found.\n")
```

```python
# Step 1: Create a new column containing the count of languages used in each movie
a['Language_Count'] = a['Language'].apply(lambda x: len(x.split('/')))

# Step 2: Filter the DataFrame to include only movies with 1, 2, or 3 languages
filtered_movies = a[a['Language_Count'].isin([1, 2, 3])]

# Step 3: List the movies based on the count of languages used
for count in [1, 2, 3]:
    movies_with_count = filtered_movies[filtered_movies['Language_Count'] == count][['Title', 'Language']]
    if not movies_with_count.empty:
        print(f"Movies with {count} language(s):")
        # Calculate total number of languages for each movie
        movies_with_count['Total_Languages'] = movies_with_count['Language'].apply(lambda x: len(x.split('/')))
        # Display the table with title, languages, and total number of languages
        print(movies_with_count.to_string(index=False))
        print()
    else:
        print(f"No movies with {count} language(s) found.\n")
```
